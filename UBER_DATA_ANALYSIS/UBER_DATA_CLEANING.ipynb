**FILLING NULL VALUES IN FARE AMOUNT**

```python
# Calculate the average fare per km from available data
df['Fare_per_km'] = df['Fare_Amount'] / df['Distance_km']
avg_fare_per_km = df['Fare_per_km'].mean()

# Fill missing Fare_Amount using Distance_km * avg_fare_per_km
df['Fare_Amount'] = df.apply(lambda row: row['Distance_km'] * avg_fare_per_km if pd.isna(row['Fare_Amount']) else row['Fare_Amount'], axis=1)

df.drop(columns=['Fare_per_km'], inplace=True)  # Drop the helper column
df.head()
```

**FILLING NULL VALUES IN PAYMENT TYPE**

```python
df['Payment_Type'] = df['Payment_Type'].fillna(method='bfill')
```

**FILLING NULL VALUES IN DRIVER RATING**

```python
df['Driver_Rating'] = df.groupby('Vehicle_Type')['Driver_Rating'].transform(lambda x: x.fillna(x.mean()))
```
